##Additional Promotion


```python
def check_promo_month(row):
    month2str = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',
                 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}
    try:
        months = (row['PromoInterval'] or '').split(',')
        if row['Promo2Open'] and month2str[row['Month']] in months:
            return 1
        else:
            return 0
    except Exception:
        return 0

def promo_cols(df):
    # Months since Promo2 was open
    df['Promo2Open'] = 12 * (df.Year - df.Promo2SinceYear) +  (df.WeekOfYear - df.Promo2SinceWeek)*7/30.5
    df['Promo2Open'] = df['Promo2Open'].map(lambda x: 0 if x < 0 else x).fillna(0) * df['Promo2']
    # Whether a new round of promotions was started in the current month
    df['IsPromo2Month'] = df.apply(check_promo_month, axis=1) * df['Promo2']
```

```python
def promo_cols(df):
    # Fill missing values in Promo2SinceYear and Promo2SinceWeek with a default value
    df['Promo2SinceYear'] = df['Promo2SinceYear'].fillna(df['Year'])  # Assume promotion started the same year if missing
    df['Promo2SinceWeek'] = df['Promo2SinceWeek'].fillna(1)  # Assume promotion started from the first week if missing

    # Calculate Months since Promo2 was open
    df['Promo2Open'] = 12 * (df['Year'] - df['Promo2SinceYear']) +  (df['WeekOfYear'] - df['Promo2SinceWeek'])*7/30.5
    df['Promo2Open'] = df['Promo2Open'].map(lambda x: 0 if x < 0 else x).fillna(0) * df['Promo2']

    # Whether a new round of promotions was started in the current month
    df['IsPromo2Month'] = df.apply(check_promo_month, axis=1).fillna(0) * df['Promo2']

# Apply the function to your DataFrames
promo_cols(merged_df)
promo_cols(merged_test_df)
```
